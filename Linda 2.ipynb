```python
from datetime import datetime
#day_of_year = datetime.now().timetuple().tm_yday
datetime(2018,10,5).timetuple().tm_yday
```

```python
def pieliktQ(row):
    yDay=row["Day of year"]
    q=row["Q - Flow rate (l/s)"]
    
    if 74 <= yDay <= 120 and not pd.isna(row["compensated waterlevel (cm)"]):
        if pd.isna(q):
            q=15000/(24*60*60)
        else:
            q=q+15000/(24*60*60)
    elif 121<=yDay<=278 and not pd.isna(row["compensated waterlevel (cm)"]):
        
        if pd.isna(q):
            q=30000/(24*60*60)
        else:
            q=q+30000/(24*60*60)
    
    return q
```

15 marts (74.diena gadā) līdz 30 (120.diena) aprīlis = 7 kubi nost, no 1. maija (121.diena) līdz 5. oktobrim (278.diena) 1 nost - 2018 un 2019 gadi, visticamāk, ka arī 2020.

```python
df_main["q_adjusted (l/s)"]=df_main.apply(pieliktQ, axis=1)
```
