```python
import pandas as pd
```

```python
df=pd.read_csv("mezaciruli.csv", parse_dates=["Date"], index_col=0, dayfirst=False)
```

```python
df
```

```python
df.loc["2015"][["1_N kop","2_N kop"]].plot()
```

```python
df["N_diff"]=df["1_N kop"]-df["2_N kop"]
```

```python
df["N_diff"].plot(grid=True)
```

```python
import seaborn as sns
```

```python
df
```

```python
df["Month"]=df.index.month
```

```python
sns.boxplot(data=df, x="Month", y="2_N kop", palette="Set3")
```

```python
s=df.apply(lambda row: 
           (row["1_N kop"]-row["2_N kop"])/row["1_N kop"]*100 
           if row["1_N kop"]>row["2_N kop"] 
           else 
           (row["1_N kop"]-row["2_N kop"])/row["2_N kop"]*100, 
           axis=1)
```

```python
df["retention N kop"]=s
```

```python
df
```

```python
sns.boxplot(data=df.loc["2016":"2020"], x="Month", y="retention N kop", palette="Set3")
```

```python
df=df.asfreq(freq="1D")
```
