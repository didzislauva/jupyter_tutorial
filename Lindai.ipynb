```python
s=df.apply(lambda row: 
           (row["1_N kop"]-row["2_N kop"])/row["1_N kop"]*100 
           if row["1_N kop"]>row["2_N kop"] 
           else 
           (row["1_N kop"]-row["2_N kop"])/row["2_N kop"]*100, 
           axis=1)
```

```python
df["retention N kop"]=s
```

```python
def efektivizators(row,name1,name2):

    if row[name1]>row[name2]:
        return (row[name1]-row[name2])/row[name1]*100
    elif row[name1]<row[name2]:
        return (row[name1]-row[name2])/row[name2]*100
    else:
        return None


    
```

```python
names1=list(df.columns)[1:8]
names2=list(df.columns)[8:15]


for enum, each in enumerate(names1):
    paramName="eff_"+names1[enum][2:]
    print(paramName)
    df[paramName]=df.apply(efektivizators,args=(names1[enum], names2[enum]), axis=1)
    
df

```
