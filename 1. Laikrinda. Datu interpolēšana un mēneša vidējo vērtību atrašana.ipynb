```python
import pandas as pd
```

Vispirms ir jāimportē bibliotēka, ar kuru strādāsim - `pandas`. Ir pieņemts, ka to saīsina uz `pd`

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)    
```

Norādīsim, ka, izsaucot tabulu, vēlamies redzēt visus datus

```python
df = pd.read_csv("dati.csv",parse_dates=['datums'], dayfirst=True, index_col=0)
```

No iepriekšsagatavota faila `dati.csv` iesauksim datus tabulā ar nosaukumu `df` (no vārda - dataframe), norādīsim, ka pirmā kolonna ar laikrindu būs datu tabulas indekss, un to, ka laikrindas formāts ir `diena.mēnesis.gads`

```python
df
```

Iesauktie dati tagad izskatās šādi.

```python
df_avg = df.resample('D').mean()
```

Tā kā mērījumi ir piefiksēti kaut kādos konkrētos laika brīžos, ir iespējams pētīt šādu **laikrindu** un, piemēram, atrast dienas vidējo (vai jebkādu citu statistisku) vērtību. Izveidosim jaunu tabulu `df_avg`, kas saturēs dienas vidējās vērtības.

Jāievēro, kā norisinās darbs ar tabulu - tabulai `df` tiek veikta tās pārrēķināšana pēc dienas `resample('D')` vidējās `mean()` vērtības un no šīs tiek izveidota jauna tabula `df_avg`

```python
df_avg
```

Apskatām iegūto tabulu un redzam, ka jaunā tabula `df_avg` ir veidota katrai dienai. Pie tam dienās, kurās mērījumi nav veikti, ir norādīts, ka mērījumu nav - `NaN`

```python
df_avg_m1 = df.resample('M').mean()
```

No **oriģinālās** mērījumu tabulas `df` varam izveidot tabulu `df_avg_m1`, kura saturēs  **mēneša vidējās** vērtības. 

Tā kā mēnešu vidējās vērtības var atrast arī no **dienas vidējo** mērījumu tabulas `df_avg`, tad salīdzināsim rezultātus, pielietojot abus paņēmienus. (Tāpēc mēneša vidējo tabulu nosaukumā galā pievienots kārtas skaitlis)

```python
df_avg_m1
```

```python
df_avg_m2=df_avg.resample('M').mean()
```

Un otrais paņēmiens - veidojot mēneša vidējo mērījumu tabulu `df_avg_m2` no dienas vidējās tabulas `df_avg`

```python
df_avg_m2
```

Salīdzinot **2016. gada jūnija** vidējo vērtību abos paņēmienos, var redzēt, ka tās atšķiras

```python
df_avg_m2.interpolate(inplace=True, method='linear')
```

Ja situācija atļauj un ir nepieciešams interpolēt vērtības starp esošajiem mērījumiem, to var paveikt dažādās formās un veidos - viens no tiem ir interpolēt lineāri.
Inerpolējot var norādīt dažādus parametrus. Šeit esmu izmantojis divus
1. `inplace=True` - tiek pārveidota esošā tabula, nevis veidota jauna (Šeit vērts piedomāt, ja vēlas saglabāt informāciju par mākslīgi ģenerētajiem datiem un oriģinālajiem mērījumiem)
2. `method='linear'` - interpolācijas metode ir ... lineāra ;)

```python
df_avg_m2
```

### Grafiku veidošana


Interpolētos datus attēlosim grafikā. Tā kā ir divas kolonnas `x1` un `x2`, tad abas norādīsim atsevišķos grafikos, jo to vērtības pa `y` asi atšķiras par vienu kārtu.

```python
cols=["x1","x2"]
```

```python
grafiki=df_avg_m2[cols].plot(marker=".", alpha=0.5, color=["orange","blue"], 
                          linestyle=None, figsize=(11,9),subplots=True, grid=True)
for grafiks in grafiki:
    grafiks.set_ylabel("vērtība")
```
